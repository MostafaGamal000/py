```python
import requests
import bs4

url = "https://cairosales.com/en/microwaves/?n=117"
page = requests.get(url)

soup = bs4.BeautifulSoup(page.content, "html.parser")
list = soup.find_all('div', {'class': 'product-container'})

print(len(list))
for item in list:
    print(item.find('a', {'class': 'product-name'}).text)
    if item.find('span', {'class': 'price product-price'}) == None:
        print("kkkkkkkkkkkkkkkkkkkkkk")
    else:
        print(item.find('span', {'class': 'price product-price'}).text[4:])
        print('\n\n')
```
